```python

# Helper Function to clean the ingredients list
def clean_ingredients(ingredient_list):
    # Convert each ingredient to lowercase
    cleaned_ingredients = [ingredient.lower() for ingredient in ingredient_list]
    # Remove unwanted characters using regular expressions
    cleaned_ingredients = [re.sub(r'[^a-z\s]', '', ingredient) for ingredient in cleaned_ingredients]
    # Remove extra whitespace
    cleaned_ingredients = [re.sub(r'\s+', ' ', ingredient) for ingredient in cleaned_ingredients]
    # Remove leading/trailing whitespace
    cleaned_ingredients = [ingredient.strip() for ingredient in cleaned_ingredients]
    return cleaned_ingredients
```

```python
# Define a list of unwanted words
unwanted_words = ['granulated','chopped','unsalted','finely','diced','low','whole','sodium','grated','black','lowfat','peeled','nonfat','ground','fresh','freshly','paste','boneless','skin','large','purple',]

# Function to remove unwanted words from a string
def remove_unwanted_words(text):
    # Create a regular expression pattern to match unwanted words
    pattern = r'\b(?:{})\b'.format('|'.join(map(re.escape, unwanted_words)))
    # Use the pattern to replace unwanted words with an empty string
    processed_text = re.sub(pattern, '', text, flags=re.IGNORECASE)
    return processed_text

# Apply the function to the 'text' column
d_excel1['ingredients'] = d_excel1['ingredients'].apply(remove_unwanted_words)
```

```python
# Define ingredients to remove
ingredients_to_remove = ['salt', 'water']


# Function to remove specified ingredients from a string
def remove_specific_ingredients(ingredient_string):
    ingredients = [ingredient.strip().strip("'") for ingredient in ingredient_string[1:-1].split(',')]
    filtered_ingredients = [ingredient for ingredient in ingredients if ingredient.lower() not in ingredients_to_remove]
    return filtered_ingredients

# Apply the function to the 'ingredients' column
d_excel1['ingredients'] = d_excel1['ingredients'].apply(remove_specific_ingredients)
```

__Performing undersampling and oversampling on dataframe to balance the dataset__

```python
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import LabelEncoder
from collections import Counter
```

```python
# Count the class distribution before any resampling
print("Class distribution before resampling:")
print(Counter(d_excel1['country']))

# Decide the desired class count
desired_class_count = 2000

# Create a dictionary to track which resampling method to apply for each class
resampling_methods = {}

# Iterate through unique cuisines
for cuisine in d_excel1['country'].unique():
    class_count = Counter(d_excel1[d_excel1['country'] == cuisine]['country'])[cuisine]

    # Calculate the number of samples to oversample or undersample
    if class_count < desired_class_count:
        resampling_methods[cuisine] = ('oversample', desired_class_count)
    elif class_count > desired_class_count:
        resampling_methods[cuisine] = ('undersample', desired_class_count)
    else:
        resampling_methods[cuisine] = ('none', desired_class_count)

# Apply resampling methods
resampled_data = pd.DataFrame()

for cuisine, (method, count) in resampling_methods.items():
    if method == 'oversample':
        # Oversample by duplicating samples
        resampled_data = pd.concat([
            resampled_data,
            d_excel1[d_excel1['country'] == cuisine].sample(count, replace=True, random_state=42)
        ])
    elif method == 'undersample':
        # Undersample to the desired class count
        resampled_data = pd.concat([
            resampled_data,
            d_excel1[d_excel1['country'] == cuisine].sample(count, random_state=42)
        ])
    else:
        # No resampling needed
        resampled_data = pd.concat([resampled_data, d_excel1[d_excel1['country'] == cuisine]])

# Count the class distribution after resampling
print("Class distribution after resampling:")
print(Counter(resampled_data['country']))
```

__Performing train test split on dataset__

```python
#Imports the required libraries from scikit-learn for text feature extraction, transformation, classification, and creating a pipeline for combining these steps.
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
#creating a dictionary where each unique ingredient is assigned an index for NLP tasks, including text representation and feature extraction.
vocabulary = {}
for ingredient, i in zip(all_ingredients, range(len(all_ingredients))):
    vocabulary[ingredient] = i
```

```python
#Function returns a list of strings, where each string represents a concatenation of ingredients from the 'ingredients' column of the dataframe
def stringify_ingredients(df):
    return [' '.join(ingredients) for ingredients in resampled_data1['ingredients']]
```

```python
str_train_ingredients= stringify_ingredients(resampled_data1)
#str_train_ingredients2
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(str_train_ingredients, resampled_data1['country'], test_size=0.2, random_state=42)
```

__1.SGD Classifier__, which is a linear classifier based on the stochastic gradient descent algorithm.

```python
from sklearn.linear_model import SGDClassifier 
# Define the hyperparameters
param_grid_sgd = {
    'clf__alpha': [0.0001, 0.001, 0.01, 0.1],  # a range of alpha values
    'clf__loss': ['hinge', 'log', 'modified_huber'],  # different loss functions
    'clf__penalty': ['l1', 'l2', 'elasticnet'],  # penalty terms
}


# Create an SGDClassifier
sgd_classifier = SGDClassifier(max_iter=1000)

# Create a pipeline
cuisine_classifier_sgd = Pipeline([
    ('vect', CountVectorizer(vocabulary=vocabulary)),
    ('tfidf', TfidfTransformer()),
    ('clf', sgd_classifier),
])

# Perform hyperparameter tuning using GridSearchCV
grid_search_sgd = GridSearchCV(cuisine_classifier_sgd, param_grid_sgd, cv=5)
grid_search_sgd.fit(X_train, y_train)

# Print the best hyperparameters found by GridSearchCV
print("Best hyperparameters (SGD):", grid_search_sgd.best_params_)

# Now we can use the best classifier for predictions
best_classifier_sgd = grid_search_sgd.best_estimator_

# Fit the classifier on the training data
best_classifier_sgd.fit(X_train, y_train)

# Evaluate the model on the training set
train_accuracy_sgd = best_classifier_sgd.score(X_train, y_train)
print(f"SGD Classifier Training Accuracy: {train_accuracy_sgd:.2f}")

# Evaluate the model on the testing set
test_accuracy_sgd = best_classifier_sgd.score(X_test, y_test)
print(f"SGD Classifier Test Accuracy: {test_accuracy_sgd:.2f}")

# Get predictions on the testing set
predicted_test_sgd = best_classifier_sgd.predict(X_test)
```

```python
# Helper function to draw the distribution of ingredients in a data frame
def draw_distribution_ingredients(df, ax=None):
    # Helper function to draw the distribution of ingredients in a data frame
    get_list_ingredients(df).sort_values(by='Number of apparition', ascending=False).head(20).plot(x='ingredients', y='Number of apparition', kind='barh', ax=ax)
    ax.set_title('Distribution of the ' + df['country'].iloc[0] + ' cuisine ingredients', fontsize=15)
    ax.tick_params(direction='out', length=6, width=4)
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(15)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(15)
    ax.set_ylabel('Ingredients', fontsize=16)
    return ax

def get_list_ingredients(df):
    ''' Helper function to get the list of ingredients from a DataFrame'''
    possible_ingredients = {}
    for a in df['ingredients']:
        for ing in a:
            possible_ingredients[ing] = 1 if (not ing in possible_ingredients) else possible_ingredients[ing] + 1
    ing_data = np.array([(key, possible_ingredients[key]) for key in possible_ingredients.keys()])
    df_ing = pd.DataFrame(data={'ingredients': ing_data[:, 0], 'Number of apparition': ing_data[:, 1].astype(np.int32)})
    return df_ing

# Suppress warnings
warnings.filterwarnings("ignore", category=UserWarning)

f, ax = plt.subplots(5, 4, figsize=(25, 55))
f.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=1)

unique_countries = resampled_data1['country'].unique()

for i, country in enumerate(unique_countries):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        draw_distribution_ingredients(resampled_data1[resampled_data1.country == country], ax=ax[i // 4, i % 4])
        ax[i // 4, i % 4].tick_params(axis='both', labelsize=15)

# Hide any remaining empty subplots
for i in range(len(unique_countries), 20):
    f.delaxes(ax[i // 4, i % 4])

# Show the plots
plt.show()
```

_Based on geographical coordinates_

```python
#geographical coordinates for the given countries (latitude, longitude)
coordinates = {
    'brazilian': (-14.235004, -51.92528),
    'british': (55.378051, -3.435973),
    'cajun_creole': (30.973377, -91.429909),
    'chinese': (35.86166, 104.195397),
    'filipino': (12.879721, 121.774017),
    'french': (46.603354, 1.888334),
    'greek': (39.074208, 21.824312),
    'indian': (20.593684, 78.96288),
    'irish': (53.349805, -6.26031),
    'italian': (41.87194, 12.56738),
    'jamaican': (18.109581, -77.297508),
    'japanese': (36.204824, 138.252924),
    'korean': (35.907757, 127.766922),
    'mexican': (23.634501, -102.552784),
    'moroccan': (31.791702, -7.09262),
    'russian': (61.52401, 105.318756),
    'southern_us': (37.09024, -95.712891),
    'spanish': (40.463667, -3.74922),
    'thai': (15.870032, 100.992541),
    'vietnamese': (14.058324, 108.277199)
}

# Calculate geographical distances
distances = np.zeros((len(coordinates), len(coordinates)))

for i, (cuisine1, (lat1, lon1)) in enumerate(coordinates.items()):
    for j, (cuisine2, (lat2, lon2)) in enumerate(coordinates.items()):
        # Use Haversine formula to calculate distance
        R = 6371  # Earth's radius in km
        dlat = np.radians(lat2 - lat1)
        dlon = np.radians(lon2 - lon1)
        a = np.sin(dlat/2) * np.sin(dlat/2) + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon/2) * np.sin(dlon/2)
        c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
        distances[i, j] = R * c

# Perform hierarchical clustering
linkage_matrix = linkage(distances, method='ward')
# Create a larger figure
plt.figure(figsize=(10, 10))

# Create a dendrogram with larger font size
dendrogram(linkage_matrix, labels=list(coordinates.keys()), leaf_font_size=14)
plt.title('Hierarchical Clustering Dendrogram')
# Rotate the x-axis labels (country names)
plt.xticks(rotation=90)
plt.show()
```
